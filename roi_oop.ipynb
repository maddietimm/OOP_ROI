```python
class ReturnOnInvestment():

    def __init__(self, monthlyRentalIncome, monthlyExpenses):
        self.monthlyRentalIncome = monthlyRentalIncome
        self.monthlyExpenses = monthlyExpenses

    def income(self):
        while True:
            self.monthlyRentalIncome = input("What will you be charging for rent? Do not use any $, only numbers. ")
            if int(self.monthlyRentalIncome) > 0:
                self.misc = input("Do you have any other money coming from this unit? Laundry, storage, pet rent, etc. If so, type in that total. Do not use any '$', only numbers. Type 0 if no expense. ")
                if int(self.misc) >= 0:
                    self.totalMonthlyIncome = int(self.monthlyRentalIncome) + int(self.misc)
                    if int(self.totalMonthlyIncome) > 0: 
                        print(f"Your monthly income is ${self.totalMonthlyIncome}.")
                        break
                    else:
                        print("We are here to calculate your rental ROI, if you do not have any income, this will be tough!")
        
    def expences(self):
        while True:
            self.monthlyExpenses = input("What are your monthly expenses for this property? Insurance, HOA's, anything you pay for this unit. Do not use any '$', only numbers. ")
            if int(self.monthlyExpenses) >= 0:
                return self.monthlyExpenses
                break
    
    def cashFlow(self):
        while True:
            self.myCashFlow = float(self.totalMonthlyIncome) - float(self.monthlyExpenses)
            if float(self.myCashFlow) >= 0:
                print(f"${float(self.myCashFlow)}0 is what you have for monthly income after expenses.")
                break

    def cashOnCash(self, totalInvestment, annualCashFlow):
        while True:
            self.downPayment = input("What was your down payment on this unit? ")
            if float(self.downPayment) >= 0:
                self.closingCosts = input("What did you pay for closing costs for this unit? ")
                if float(self.closingCosts) >= 0:
                    self.miscCost = input("What other costs did you pay for this unit? Rehab budget, appliances, etc. ")
                    if float(self.miscCost)>=0:
                        self.annualCashFlow = float(self.totalMonthlyIncome)*12
                        if float(self.annualCashFlow) >= 0:
                            self.totalInvestment = float(self.downPayment) + float(self.closingCosts) + float(self.miscCost)
                            if self.totalInvestment>=0:
                                self.roi = ((int(self.annualCashFlow)) / (int(self.totalInvestment)) * 10)
                                print(f"Great! Your ROI is {float(self.roi)}%. We hope to help you in the future! ")
                                break
            else:
              False
        
deeperPockets = ReturnOnInvestment("","")

def run():
    while True:
        welcome = input("Welcome! Would you like to get started with your ROI calculations? ")
        if welcome.lower() == 'yes':
            deeperPockets.income("","","")
            while True:
              deeperPockets.expences()
              while True:
                deeperPockets.cashFlow()
                while True:
                  deeperPockets.cashOnCash("","")     
                  break
                break 
              break
            break         
        else:
            print("We hope we can help you in the future! See you later!")
            break
run()
```
